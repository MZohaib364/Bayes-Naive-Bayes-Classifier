````python
# YOUR CODE HERE
# raise NotImplementedError()

class GaussianBayes:
    ''' Implements the Gaussian Bayes For Classification... '''
    def __init__(self):
        self.class_labels = {}
        self.prior_probs = {}
        self.mean_vectors = {}
        self.std_vectors = {}
        
        pass
    
    def train(self, X, Y):
        ''' Train the multiclass (or Binary) Bayes Rule using the given 
            X [m x n] data matrix and Y labels matrix'''
        
        # YOUR CODE HERE
        
        self.class_labels = np.unique(Y)
        
        
        self.prior_probs = {class_label: 0 for class_label in self.class_labels}
        
        class_wise_data = {class_label: [] for class_label in self.class_labels}       
        for i in range(len(Y)):
            class_wise_data[Y[i]].append(X[i])
        
        self.mean_vectors = {class_label: [] for class_label in self.class_labels}
        self.std_vectors = {class_label: [] for class_label in self.class_labels}
        
        for class_label in self.class_labels:
            self.prior_probs[class_label] = (len(class_wise_data[class_label]) / len(Y))
            self.mean_vectors[class_label].append(np.mean(class_wise_data[class_label], axis=0))
            self.std_vectors[class_label].append(np.std(class_wise_data[class_label], axis=0))
            
#         raise NotImplementedError()
        
    def test(self, X):
        
        ''' Run the trained classifiers on the given set of examples 
            For each example, you should return probability and its assigned class
            Input: X of m x d
            Output:
            pclasses: predicted class of each example
            probability: probability of each example falling in that predicted class...
        '''
        
        nexamples, nfeatures=X.shape
        res = []
        pclasses=[]
        
        
        # YOUR CODE HERE
        
        def gaussian_pdf(x, m, std):
            return (1 / (std * np.sqrt(2*np.pi))) * (np.exp(-0.5 * ((x-m) / std)**2))
        

        for i in range(nexamples):
            c_wise_prob = []
            for class_label in self.class_labels:
                indv_feat_probs = []     # [P(X1|c), P(X2|c)......]
                for j in range(nfeatures):
                    indv_feat_probs.append(gaussian_pdf(np.array(X[i][j]), self.mean_vectors[class_label][0][j], self.std_vectors[class_label][0][j]))
                c_wise_prob.append(np.prod(indv_feat_probs) * self.prior_probs[class_label])

#             print("c_wise_prob", c_wise_prob)
            max_index = np.argmax(c_wise_prob)
            res.append(c_wise_prob[max_index])
            pclasses.append(self.class_labels[max_index])
        
#         raise NotImplementedError()
        
        return np.array(pclasses), res
    
    def predict(self, X):
        return self.test(X)[0]    
        
````

# K-Fold Cross Validation

Since our dataset is not too big, to ensure that we are reporting true picture of our classifier, 
we will have to cross validate and report the mean accuracy across the folds to reflect the true 
picture of our classifier.

````python
# Now lets cross validate, and test the result...
nfolds=4
folds=t.generate_folds(X,Y,nfolds)

#now lets train and test on these folds...
totacc=[]
for k in range(nfolds):
    gb=GaussianBayes()
    gb.train(folds[k][0][:,2:],folds[k][1])
    pclasses,res=gb.test(folds[k][2][:,2:])
    acc=np.sum(pclasses==folds[k][3])/float(folds[k][3].shape[0])
    print ("[Info] Fold {} Accuracy = {}".format(k+1, acc))
    totacc.append(acc)

print (totacc, 'Mean Accuracy =', np.mean(totacc))
````
